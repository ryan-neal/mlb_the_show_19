```python
import pandas as pd
```

```python
steamers = pd.read_csv("Steamer_projections.csv").set_index(["playerid", "Name"])
fangraphs = pd.read_csv("Fangraph_projections.csv").set_index(["playerid", "Name"])
zips = pd.read_csv("Zips_projections.csv").set_index(["playerid", "Name"])
```

```python
quantitative = [f for f in steamers if steamers.dtypes[f] != object]    
```

```python
steamers_quant = steamers[quantitative]
fangraphs_quant = fangraphs[quantitative]
zips_quant = zips[quantitative]
```

```python
steamers_quant
```

```python
df = steamers_quant.add(fangraphs_quant, fill_value=0)
```

```python
df = df.add(zips_quant, fill_value=0)
```

```python
df = df/3
```

```python
df = df.round(decimals=1)
```

```python
df = df.sort_values(by="WAR")
```
